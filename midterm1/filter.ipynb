```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Load the image
img = cv2.imread('Lenna.png')
```

```python
# zero mean

# Define the filter size
filter_size = (5, 5)

# Compute the local mean of each pixel
kernel = np.ones(filter_size, dtype=np.float32) / (filter_size[0] * filter_size[1])
local_mean = cv2.filter2D(img, -1, kernel)

# Subtract the local mean from each pixel
output_img = img - local_mean

# Display the input and output images
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(10, 5))

ax1.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax1.set_title('Input Image')

ax2.imshow(cv2.cvtColor(output_img, cv2.COLOR_BGR2RGB))
ax2.set_title('Output Image')

plt.show()

# Save the output image
cv2.imwrite('output_image.png', output_img)
```
